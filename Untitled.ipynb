```python
#scatter plot
plt.figure(figsize=[12,7],)
plt.scatter(data['Age'],data['Fare'],color='red')
plt.xlabel("Age of Passengers")
plt.ylabel("Age-Fare")
plt.show 
```

```python
#grouping adta by different colors
color=[]
for Sex in data['Sex']:
    if Sex=='male':
        color.append('blue')
    else:
            color.append('pink')
plt.figure(figsize=[12,7],)            
plt.scatter(data['Age'],data['Fare'],c=color)
plt.xlabel('Age of Passengers')
plt.ylabel('Fare')
plt.title('Age-Fare')
plt.show
```
